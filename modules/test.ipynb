## スクレイピングに失敗した馬の再スクレイピング

```python
import glob
import os
from io import StringIO

import pandas as pd
from tqdm.notebook import tqdm

import my_snipets

year = "2020"
files = glob.glob(f"../../../data/rawData/raceResults/{year}*")
files.sort()

raceResults = pd.DataFrame()
for file in tqdm(files):
    dftmp = pd.read_pickle(file)
    raceResults = pd.concat([raceResults,dftmp])

horseIDList = raceResults["horseID"].unique().tolist()
horseIDList.sort()

misshorseIDList = []
for horseID in tqdm(horseIDList):
    if os.path.isfile(f"../../../data/rawData/horse/{horseID}.pickle") == False:
        try:
            with open(f"../../../data/html/horse/{horseID}.txt","r",encoding="utf-8") as f:
                html = f.read()

            #htmlから過去戦績のデータを抽出
            contents = pd.read_html(StringIO(html))
            if len(contents) < 3 :
                horseResults = pd.DataFrame(columns=my_snipets.getHorseResultsColumns())
            elif contents[2].columns[0] == "日付":
                horseResults = contents[2]
            else:
                horseResults = contents[3]
            
            #データフレームをpickle形式で保存
            horseResults.to_pickle(f"../../../data/rawData/horse/{horseID}.pickle")
        except IndexError:
            misshorseIDList.append(horseID)
```

```python
#raceCardDFのhorseIDのデータを作成
from makeRawData import makeHorseData
MHD = makeHorseData(misshorseIDList)
MHD.getHorseHTML(skipResult=False, skipPED=False)
MHD.makeHorseRawData()
```

```python
import datetime

import pandas as pd

from makeRawData import makeHorseData
from preprocessing import preprocessingRaceCard
from prediction import makeRaceCardData, predLGBM, predNN, predDelivery

tgtyear = 2025
tgtmonth = 9
tgtday = 15

mRCD = makeRaceCardData(datetime.date(tgtyear, tgtmonth, tgtday))
mRCD.makeRaceDateURLList()
mRCD.makeRaceURLList()

runtimeList = []
for tgt in mRCD.urlDF["datetime"]:
    runtime = tgt + datetime.timedelta(minutes=-3)
    runtimeList.append(runtime)

for idx, runtime in enumerate(runtimeList):
    returnflg = 0
    while returnflg == 0:
        now = datetime.datetime.now()
        if now.hour == runtime.hour and now.minute == runtime.minute :
            raceID = mRCD.screpingRaceCardHTML(raceNo=idx)
            raceCardDF = mRCD.makeRaceCardDataFrame(raceNo=idx)

            MHD = makeHorseData(raceCardDF["horseID"].tolist())
            MHD.getHorseHTML(skipResult=False)
            MHD.makeHorseRawData()

            pRC = preprocessingRaceCard(raceCardDF)
            pRC.raceCardPreprocess()
            pRC.makedataHorseResults()
            pRC.horseResultsPreprocess()
            pRC.makeVarFromHorseResults()
            forScoringDF = pRC.forScoringDFPreprocess()

            # lightGBMによる予測
            pLGBM = predLGBM("LGBM_1", forScoringDF, raceCardDF)
            pLGBM.loadModel()
            raceCardDF = pLGBM.pred()

            # NNによる予測
            pNN = predNN("NN_1", forScoringDF, raceCardDF)
            pNN.loadModel()
            raceCardDF = pNN.pred()

            pDeli = predDelivery(raceID, raceCardDF)
            pDeli.makeRaceInfoText()
            pDeli.makePredText(5)
            pDeli.delivaryPredLine()
            returnflg = 1
        else:
            time.sleep(30)
```
